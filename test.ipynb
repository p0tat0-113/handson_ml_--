```python
import tensorflow as tf

# 1) Growth 모드 활성화
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)

# 2) 아주 단순한 덧셈 연산
with tf.device('/GPU:0'):
    a = tf.constant([[1., 2.], [3., 4.]])
    b = tf.constant([[5., 6.], [7., 8.]])
    c = tf.add(a, b)
print("Result:", c)
```
